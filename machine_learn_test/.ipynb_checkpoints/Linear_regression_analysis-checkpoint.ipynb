```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

%matplotlib inline
data = pd.read_csv('dataset.csv')
data
```

```python
data=np.array(data)
train_data, test_data, train_target, test_target = train_test_split(data[:,1:8],data[:,8],test_size=0.3,random_state=0)
# ----------------------------
# print("train_data")
# print(train_data.shape)
# print("train_target")
# print(train_target.shape)
# print("test_data")
# print(test_data.shape)
# print("test_target")
# print(test_target.shape)
```

```python
#----构建模型----
reg = linear_model.LinearRegression()
#----训练模型----
reg.fit(train_data, train_target)
#----评估模型----
# print(reg.coef_)        # 斜率
# print(reg.intercept_)   # 截距
print(reg.score(train_data, train_target))
#----目標預測----
data_target_pred = reg.predict(test_data)
# print(data_target_pred)
# print(data_target_pred.shape)

plt.ylabel('Test Set')
plt.xlabel('Predicted values')
plt.scatter(data_target_pred, test_target,  color='blue')
plt.show()
# print(test_target)
```

```python
# Explained variance score: 1 is perfect prediction 
#  R2 決定係數（擬合優度）
# 模型越好：r2→1
# 模型越差：r2→0

print("real value of y_test[1]:%.2f" %test_target[1] + "-> the predict:[ %.9f ]" %data_target_pred[1])
print("real value of y_test[2]:%.2f" %test_target[2] + "-> the predict:[ %.9f ]" %data_target_pred[2])

print("r_square score: %.2f" % r2_score(test_target, data_target_pred))
# print("r_square score: %.2f" % r2_score(data_target_pred))
```

```python
# The mean squared error  
# 均方差
# 檢視殘差平方的均值(mean square error,MSE) 
print("Mean squared error: %.15f " % float(mean_squared_error(test_target, data_target_pred)*100) )
```
