```python
import numpy as np
import pandas as pd
import keras
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.cluster import KMean
%matplotlib inline
data = pd.read_csv('dataset.csv')
data
```

```python
data_feature = data.iloc[:,1:9]
data_feature
```

```python
cluster = KMeans(n_clusters=3)
cluster.fit(data_feature)
plt.xlabel('cluster group')
plt.ylabel('count')
bins = np.arange(0, 5, 0.5) # fixed bin size
plt.hist(cluster.labels_, bins=bins)
plt.show()
```

```python
plt.scatter(cluster.labels_,data_feature['Chance of Admit '])
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,10))
sn.boxplot(cluster.labels_,data_feature['Chance of Admit '])
```
